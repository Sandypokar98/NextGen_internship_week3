## Bar Plot

```python
x=["Python","C","C++","Java"]
y=[85,70,80,82]
c=["yellow","Blue","green","pink"]
plt.xlabel("Language",fontsize=11)
plt.ylabel("Popularity",fontsize=11)
plt.title("Bar Plot",fontsize=18)
plt.bar(x,y,width=0.4,color="c",align="center",edgecolor="Black",linewidth=2,linestyle=":",alpha=0.7,label="Popularity")
plt.legend()
```

## Scatter plot

```python
x=np.linspace(0,31.5,78)
y=np.sin(x/5)
```

```python
plt.scatter(x,y,label="sine",c="r",s=200,alpha=.6,marker="*",edgecolor="y",linewidth=1.5)
plt.xlabel("Sine",fontsize=11)
plt.ylabel("Angle",fontsize=11)
plt.title("Sine Graph",fontsize=18)
plt.legend()
```

```python
x=np.linspace(1,10,10)
y=x**3-2*x**2-30
y
```

```python
plt.scatter(x,y,label="sine",c=y,cmap="viridis",s=[400,200,150,800,350,600,150,350,600,850],edgecolor="r",linewidth=2)
plt.xlabel("Sine",fontsize=11)
plt.ylabel("Angle",fontsize=11)
plt.title("Sine Graph",fontsize=18)
plt.legend()
plt.colorbar()
```

## Histogram Plot

```python
l=np.linspace(0,max(df.price),20)
plt.hist(df.price,label="Price",color="r",edgecolor="b",bins=l,log=True)
plt.xlabel("Price",fontsize=11)
plt.ylabel("Freq",fontsize=11)
plt.title("Diamond graph",fontsize=18)
plt.axvline(df.price.median(),color="yellow",label="Median")
plt.axvline(df.price.mean(),color="green",label="Mean")
plt.legend()
```

```python
plt.hist(df.price,label="Price",color="r",edgecolor="b",bins=l,cumulative=1,rwidth=.8,orientation="horizontal")
plt.xlabel("Cum freq",fontsize=11)
plt.ylabel("Price",fontsize=11)
plt.title("Diamond graph",fontsize=18)
plt.legend()
```

```python
plt.hist(df.price,label="Price",color="r",edgecolor="b",bins=l,cumulative=-1,histtype="step")
plt.xlabel("Price",fontsize=11)
plt.ylabel("Freq",fontsize=11)
plt.title("Diamond graph",fontsize=18)
plt.legend()
plt.grid()
```

## Pie plot

```python
#plt.pie(df.color)
df["Count"]=1
df_count=df.groupby("color").count()
df_count.index
```

```python
ex=[0.15,0,0,0,0.15,0,0]
c=["red","orange","yellow","green","blue","indigo","m"]
plt.pie(df_count.Count,labels=df_count.index,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1.25,
        labeldistance=0.8,textprops={"fontsize":15})
plt.title("Diamomd Color")
plt.show()
```
